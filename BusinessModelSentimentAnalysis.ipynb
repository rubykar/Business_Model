```python
import pandas as pd
import numpy as np
import re
from nltk.corpus import stopwords
from wordcloud import WordCloud, STOPWORDS
import spacy


data_train = pd.read_csv('/content/training_data.csv')
data_test = pd.read_csv('/content/test_data.csv')
data_train.shape
```

```python
nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])
```

```python
############## Converting title and headlines into lowercase ##############
data_train['new_title'] = data_train['Title'].apply(lambda x: " ".join(x.lower() for x in x.split()))
data_train['new_headlines'] = data_train['Headline'].apply(lambda x: " ".join(x.lower() for x in x.split()))
data_train['new_title'].head()
```

```python
data_train['new_headlines'].head()
```

```python
########## To remove punctuation #######
data_train['new_title'] = data_train['Title'].str.replace('[^\w\s]','')
data_train['new_headlines'] = data_train['Headline'].str.replace('[^\w\s]','')
```

```python
data_train['new_title'].head()
```

```python
data_train['new_headlines'].head()
```

```python
def remove_emoji(text):
 emoji_pattern = re.compile("["
                        u"\U0001F600-\U0001F64F"  # emoticons
                        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                        u"\U0001F680-\U0001F6FF"  # transport & map symbols
                        u"\U0001F1E0-\U0001F1FF"  # flags 
                        u"\U00002702-\U000027B0"
                        u"\U000024C2-\U0001F251"
                        "]+", flags=re.UNICODE)
 return emoji_pattern.sub(r'', text)
data_train['new_title'] = data_train['Title'].apply(lambda x: remove_emoji(x))
data_train['new_headlines'] = data_train['Headline'].apply(lambda x: remove_emoji(x))
```

```python
data_train['new_title'].head()
```

```python
data_train['new_headlines'].head()
```

```python
################### REMOVING STOP WORDS###################
import nltk
nltk.download('stopwords')
stop = set(stopwords.words('english'))
data_train['new_title'] = data_train['new_title'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
data_train['new_headlines'] = data_train['new_headlines'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

```python
data_train['new_headlines'].head(20)
```

```python
data_train['new_title'].head(20)
```

```python
###############lemmatization using Spacy so that we can count the appearance of each word. Lemmatization removes the grammar tense and transforms each word into its original form. #######
def space(comment):
    doc = nlp(comment)
    return " ".join([token.lemma_ for token in doc])
data_train['new_headlines']= data_train['new_headlines'].apply(space)
data_train.head()
```

```python
data_train = data_train.drop(['PublishDate','Title','Source','Headline'],axis = 1)
data_train.head(20)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_selection import chi2, SelectKBest
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearnex import patch_sklearn

# Extract features from the title
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(data_train['new_title'])
# Convert the sentiment to categorical labels
bins = [-1, -0.5, 0, 0.5, 1, 2]  # add an additional bin edge for positive sentiment
labels = ['negative', 'somewhat negative', 'neutral', 'somewhat positive', 'positive']
y = pd.cut(data_train['SentimentTitle'], bins=bins, labels=labels)

# Perform feature selection
selector = SelectKBest(chi2, k=1000)
X_new = selector.fit_transform(X, y)

# Get the most important feature indices
indices = selector.get_support(indices=True)

# Get the corresponding feature names
features = vectorizer.get_feature_names_out()
important_features = [features[i] for i in indices]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, random_state=42)
print(X,y)

# Initialize the OneAPI estimator
patch_sklearn()

# from sklearnex import enable_ipp, enable_mkl  # Required to use OneAPI
# enable_ipp()
# enable_mkl()

estimator = LogisticRegression(max_iter=1000, n_jobs=-1, solver='saga')

# Train the model
estimator.fit(X_train, y_train)
print(X_train)
# Make predictions on the testing set
y_pred = estimator.predict(X_test)

# Evaluate the model performance
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
y_pred
# # Train a logistic regression model
# model = LogisticRegression()
# model.fit(X_train, y_train)

# # Make predictions on the testing set
# y_pred = model.predict(X_test)

# # Evaluate the model performance
# accuracy = accuracy_score(y_test, y_pred)
# print('Accuracy:', accuracy)
```

2. **DATA PROCESSING FOR TEST DATASET**


DO WE NEED TO PROCESS OUR TEST DATASET ?

Yes, it is important to process your test dataset before using it to evaluate your machine learning model. Data processing involves various steps such as cleaning, normalization, feature engineering, and encoding. These steps help to ensure that the test data is in the same format and range as the training data, and that it does not contain any missing values or outliers.

Processing the test dataset is necessary to make sure that the evaluation of the model is fair and accurate. If the test dataset is not processed, it may lead to inaccurate evaluation metrics and unreliable performance estimates. This can result in poor model performance when the model is deployed in real-world scenarios.

Therefore, it is recommended to apply the same data processing steps to the test dataset that were applied to the training dataset. This ensures that the test data is consistent with the training data, and that the model is evaluated on data that is representative of the real-world scenario it is intended for.
